```python
import torch
import torch.nn.functional as F

q = torch.ones(1, 1, 4, 8).bfloat16().cuda()  # q_len=4
k = torch.ones(1, 1, 8, 8).bfloat16().cuda()  # kv_len=8
v = torch.arange(64).reshape(1,1,8,8).float().bfloat16().cuda()

out = F.scaled_dot_product_attention(q, k, v, is_causal=True)
print(out[0,0])  # 看每个 q 行 attend 了哪些 k
```
